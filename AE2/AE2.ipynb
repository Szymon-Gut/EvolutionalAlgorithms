# Zbiór z promieniem równym 800

```python
r800 = DataLoader().load('r800.csv')
```

```python
e800 = Evolution(crossover_ratio=0.7, mutation_ratio=0.2,
                     crossover=mean_crossover, mutation=int_mutation,
                     fitness_func=r800.fitness_function,
                     selection_func=elite_selection, population_size=100,
                     individual_size=r800.n_rectangles)

e800.fit(iterations=150)
e800.visualise()
e800.save_history_to_csv()
```

```python
RECTANGLES
```

```python
RECTANGLES = [copy.deepcopy(r800.rectangles[i]) for i, n in enumerate(e800.best_individual) for j in range(n)]
place_in_circle(RECTANGLES, 800)
draw_rectangles(RECTANGLES, 800, e800.name)
```
